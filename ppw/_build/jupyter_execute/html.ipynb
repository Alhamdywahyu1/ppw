# Web Usage

## menggunakan dataset offline 

```python
import pandas as pd
df = pd.read_csv('/content/webuage.csv')
print("Dataset berhasil dimuat. 5 baris pertama:")
display(df)
```

### Filter Status 200


```python
df = df[df['Status'] == 200]
print("DataFrame difilter untuk 'Status' == 200")
display(df)
```

### Filter HTML Pages



```python
df = df[df['Request URI'].str.contains('html', na=False)]
print("DataFrame difilter untuk 'Request URI' yang mengandung 'html'. 5 baris pertama:")
display(df)
```

### filter get



```python
df = df[df['Request method'] == 'GET']
print("DataFrame difilter untuk 'Request method' == 'GET'")
display(df)
```

### ubah jadi tanggal dan waktu


```python
df['Request time'] = pd.to_datetime(df['Request time'], format='ISO8601', errors='coerce')
df['tanggal'] = df['Request time'].dt.date
df['waktu'] = df['Request time'].dt.time
print("Mengkonversi 'Request time' menjadi 'tanggal' dan 'waktu'.")
display(df[['Request time', 'tanggal', 'waktu']])
```

### urutkan berdasarkan ip

```python
from pandas.api.types import CategoricalDtype

# Dapatkan urutan unik IP berdasarkan kemunculan pertama
unique_ips_ordered = df['Remote host'].drop_duplicates().tolist()

# Buat tipe kategori dengan urutan yang ditentukan
ip_category_type = CategoricalDtype(categories=unique_ips_ordered, ordered=True)

# Konversi kolom 'Remote host' ke tipe kategori
df['Remote host'] = df['Remote host'].astype(ip_category_type)

# Urutkan DataFrame berdasarkan 'Remote host' yang sekarang berupa kategori
df_sorted_by_first_ip = df.sort_values(by='Remote host')

print("DataFrame diurutkan berdasarkan 'Remote host' (IP address) berdasarkan kemunculan pertama. 5 baris pertama:")
display(df_sorted_by_first_ip)
```

### membuat sesi

```python
df['Request time'] = pd.to_datetime(df['Request time'])
df = df.sort_values(by=['Remote host', 'Request time'])
df['time_diff'] = df.groupby('Remote host')['Request time'].diff()
print("Selisih waktu antara permintaan berturut-turut untuk setiap 'Remote host' telah dihitung dan disimpan dalam kolom 'time_diff'. 5 baris pertama:")
display(df.head())
```

```python
df['Request time'] = pd.to_datetime(df['Request time'])
df = df.sort_values(by=['Remote host', 'Request time'])
df['time_diff'] = df.groupby('Remote host', observed=False)['Request time'].diff()
print("Selisih waktu antara permintaan berturut-turut untuk setiap 'Remote host' telah dihitung dan disimpan dalam kolom 'time_diff'. 5 baris pertama:")
display(df.head())
```

```python
THIRTY_MINUTES = pd.Timedelta(minutes=30)
df['new_session'] = (df['time_diff'] > THIRTY_MINUTES) | (df['time_diff'].isnull())
print("Kolom 'new_session' telah dibuat berdasarkan selisih waktu 30 menit atau entri pertama.")
display(df[['Remote host', 'Request time', 'time_diff', 'new_session']].head(10))
```

```python
df['Session ID'] = df.groupby('Remote host', observed=False)['new_session'].cumsum()
df['Session ID'] = df['Session ID'].astype(str)
print("Kolom 'Session ID' telah dibuat.")
display(df[['Remote host', 'Request time', 'time_diff', 'new_session', 'Session ID']])
```

```python
import pandas as pd
import os
import glob

# Find the CSV file within the downloaded dataset directory
csv_files = glob.glob(os.path.join(dataset_path, '*.csv'))

if csv_files:
    # Assuming the first CSV file found is the primary one
    primary_csv_path = csv_files[0]
    df = pd.read_csv(primary_csv_path)
    print(f"Dataset loaded from: {primary_csv_path}")
    print("5 baris pertama dari DataFrame yang baru dimuat:")
    display(df.head())
else:
    print("Tidak ditemukan file CSV di dalam direktori dataset yang diunduh.")
```

### filter status 200



```python
df = df[df['response'] == 200]
print("DataFrame difilter untuk 'response' == 200")
display(df.head())
```

### filter url

```python
df = df[df['url'].str.contains('html', na=False)]
print("DataFrame difilter untuk 'url' yang mengandung 'html'. 5 baris pertama:")
display(df.head())
```

### filter method get



```python
df = df[df['method'] == 'GET']
print("DataFrame difilter untuk 'method' == 'GET'. 5 baris pertama:")
display(df.head())
```

### ubah tanggal dan waktu



```python
df['Request time'] = pd.to_datetime(df['time'], unit='s')
df['tanggal'] = df['Request time'].dt.date
df['waktu'] = df['Request time'].dt.time
print("Mengkonversi 'time' menjadi 'Request time' (datetime), 'tanggal' dan 'waktu'.")
display(df[['Request time', 'tanggal', 'waktu']].head())
```

### melakukan pengurutan berdasarkan orang



```python
from pandas.api.types import CategoricalDtype

# Get the unique IPs in order of their first appearance
unique_ips_ordered = df['host'].drop_duplicates().tolist()

# Create a categorical type with the defined order
ip_category_type = CategoricalDtype(categories=unique_ips_ordered, ordered=True)

# Convert the 'host' column to the categorical type
df['host'] = df['host'].astype(ip_category_type)

# Sort the DataFrame based on the 'host' column (which is now categorical)
df = df.sort_values(by='host')

print("DataFrame diurutkan berdasarkan 'host' (IP address) berdasarkan kemunculan pertama. 5 baris pertama:")
display(df.head())
```

### melihat jarak waktu dari tindakan seseorang

```python
df = df.sort_values(by=['host', 'Request time'])
df['time_diff'] = df.groupby('host', observed=False)['Request time'].diff()
print("Selisih waktu antara permintaan berturut-turut untuk setiap 'host' telah dihitung dan disimpan dalam kolom 'time_diff'. 5 baris pertama:")
display(df[['host', 'Request time', 'time_diff']].head())
```

### melakukan pembagian sesi

```python
THIRTY_MINUTES = pd.Timedelta(minutes=30)
df['new_session'] = (df['time_diff'] > THIRTY_MINUTES) | (df['time_diff'].isnull())
print("Kolom 'new_session' telah dibuat berdasarkan selisih waktu 30 menit atau entri pertama.")
display(df[['host', 'Request time', 'time_diff', 'new_session']].head(10))
```

### membuat sesi untuk tiap orang

```python
df['Session ID'] = df.groupby('host', observed=False)['new_session'].cumsum()
df['Session ID'] = df['Session ID'].astype(str)
print("Kolom 'Session ID' telah dibuat.")
display(df[['host', 'Request time', 'new_session', 'Session ID']].head(10))
```

```python
display(df)
```

### melihat ke halaman web mana saja user dalam 1 sesi

```python
df['Session_User_ID'] = df['host'].astype(str) + '_' + df['Session ID'].astype(str)

# Create the pivot table using pd.crosstab
session_page_matrix = pd.crosstab(index=df['Session_User_ID'], columns=df['url'])

# Convert counts to 0 or 1 (visited or not visited)
# Any value > 0 means the page was visited in that session, so set to 1.
# Otherwise, it was not visited, so set to 0.
session_page_matrix = session_page_matrix.applymap(lambda x: 1 if x > 0 else 0)

print("Tabel baru 'session_page_matrix' telah dibuat. 5 baris pertama:")
display(session_page_matrix.head())
```

### membuat mayoritas tindakan dari seluruh sesi milik si user


```python
# Extract the 'host' from the 'Session_User_ID' index
session_page_matrix['host'] = session_page_matrix.index.str.split('_').str[0]

# Group by 'host' and calculate the mean for each page.
# The mean will represent the proportion of sessions in which a page was visited.
user_page_matrix_majority = session_page_matrix.groupby('host').mean()

# Apply the majority rule: if the proportion is > 0.5, set to 1; otherwise, set to 0.
user_page_matrix_majority = user_page_matrix_majority.applymap(lambda x: 1 if x > 0.5 else 0)

print("Tabel baru 'user_page_matrix_majority' telah dibuat berdasarkan tindakan mayoritas di tiap sesi. 5 baris pertama:")
display(user_page_matrix_majority)
```
